### Importing Libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import Normalizer, StandardScaler
import tensorflow as tf
import keras
from tensorflow.keras.optimizers import Adam
```

### Reading csv file

```python
dataset = pd.read_csv('life_expectancy.csv')
dataset.head()
```

### Inspecting Data

```python
dataset.info()
```

```python
dataset.describe()
```

### Dropping Countries column

```python
dataset = dataset.drop('Country', axis = 1)
dataset.head()
```

### Labels and Features selection

```python
labels = dataset[['Life expectancy']]
labels.head()
```

```python
features = dataset.iloc[:,:-1]
features.head()
```

### Data Preprocessing

```python
features = pd.get_dummies(features)
features.head()
```

### Data split into training and testing sets

```python
features_train, features_test, labels_train, labels_test = train_test_split(features, labels, test_size = 0.2, random_state = 42)
```

### Standardize Numerical features

```python
numerical_features = features.select_dtypes(include = ['float64', 'int64'])
numerical_columns = numerical_features.columns
```

```python
ct = ColumnTransformer([('standardize', StandardScaler(), numerical_columns)], remainder = 'passthrough')
```

```python
features_train = ct.fit_transform(features_train)
features_test = ct.transform(features_test)
```

### Building the model

```python
my_model = tf.keras.models.Sequential()
```

```python
input = tf.keras.layers.InputLayer(input_shape = (dataset.shape[1], ))
my_model.add(input)
```

```python
hidden = tf.keras.layers.Dense(8, activation = 'relu')
my_model.add(hidden)
```

```python
output = tf.keras.layers.Dense(1)
my_model.add(output)
```

```python
my_model.summary()
```

### Initializing the optimizer and compiling the model

```python
opt = Adam(learning_rate = 0.01)
my_model.compile(loss = 'mse', metrics = ['mae'], optimizer = opt)
```

```python
my_model.fit(features_train, labels_train, epochs = 40, batch_size = 1, verbose = 1)
```

### Fit and evaluate the model

```python
res_mse, res_mae = my_model.evaluate(features_test, labels_test, verbose = 0)
```
